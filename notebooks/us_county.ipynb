## US county-level data are not recommended for use. Reporting has been inconsistent. JH may be moving to just state-level data for the US

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import pickle
from pprint import pprint as pp
```

```python
def sum_by_day(df):
    return df.groupby('day').sum().reset_index()

df_us_county = pickle.load(open('../pickles/df_c19us_county.p', 'rb'))
df = df_us_county['cases']
df = sum_by_day(df)
```

```python
plt.plot(df.day, df.cases, 'bo')
plt.title('US county-level data over time')
plt.ylabel('Cases')
plt.xlabel('Days since 2020-01-22')
plt.show()
```
